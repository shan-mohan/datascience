```python
import nltk
nltk.download('inaugural')
from nltk.corpus import inaugural
inaugural.fileids()
text_1941 = inaugural.raw('1941-Roosevelt.txt')
text_1961 = inaugural.raw('1961-Kennedy.txt')
text_1973 = inaugural.raw('1973-Nixon.txt')
```

```python
num_characters_1941 = len(text_1941)
num_words_1941 = len(nltk.word_tokenize(text_1941))
num_sentences_1941 = len(nltk.sent_tokenize(text_1941))
num_characters_1961 = len(text_1961)
num_words_1961 = len(nltk.word_tokenize(text_1961))
num_sentences_1961 = len(nltk.sent_tokenize(text_1961))
num_characters_1973 = len(text_1973)
num_words_1973 = len(nltk.word_tokenize(text_1973))
num_sentences_1973 = len(nltk.sent_tokenize(text_1973))

print("number of characters in 1941-Roosevelt:", num_characters_1941)
print("number of words in 1941-Roosevelt:", num_words_1941)
print("number of sentences in 1941-Roosevelt:", num_sentences_1961)
print("number of characters in 1961-Kennedy:", num_characters_1961)
print("number of words in 1961-Kennedy:", num_words_1961)
print("number of sentences in 1961-Kennedy:", num_sentences_1941)
print("number of characters in 1973-Nixon:", num_characters_1973)
print("number of words in 1973-Nixon:", num_words_1973)
print("number of sentences in 1973-Nixon:", num_sentences_1973)
```

2.2) Remove all the stopwords from the three speeches. Show the word count before and after the removal of stopwords. Show a sample sentence after the removal of stopwords.

```python
from nltk.corpus import stopwords
from nltk.corpus import inaugural
from string import punctuation

nltk.download('stopwords')
nltk.download('punkt')
stop_words = set(stopwords.words('english'))
stop_words_with_punctuation = stop_words.union(set(punctuation))
stop_words_with_punctuation.add('--')
```

```python
print("number of words before stopword removal in 1941-Roosevelt:", num_words_1941)
print("number of words before stopword removal in 1961-Kennedy:", num_words_1961)
print("number of words before stopword removal in 1973-Nixon:", num_words_1973)
words_1941 = nltk.word_tokenize(text_1941)
filtered_words_1941 = [word for word in words_1941 if word.lower() not in stop_words_with_punctuation]
after_stopword_count_1941 = len(filtered_words_1941)
print("Words Count after removing stopword 1941-Roosevelt:", after_stopword_count_1941)
words_1961 = nltk.word_tokenize(text_1961)
filtered_words_1961 = [word for word in words_1961 if word.lower() not in stop_words_with_punctuation]
after_stopword_count_1961 = len(filtered_words_1961)
print("Words Count after removing stopword 1961-Kennedy:", after_stopword_count_1961)
words_1973 = nltk.word_tokenize(text_1973)
filtered_words_1973 = [word for word in words_1973 if word.lower() not in stop_words_with_punctuation]
after_stopword_count_1973 = len(filtered_words_1973)
print("Words Count after removing stopword 1973-Nixon:", after_stopword_count_1973)
print("1941-Roosevelt sample:", filtered_words_1941)
```

```python
from collections import Counter
```

```python
word_counter_1941 = Counter(filtered_words_1941)
word_counter_1961 = Counter(filtered_words_1961)
word_counter_1973 = Counter(filtered_words_1973)
```

```python
print('Top 3 word in 1941-Roosevelt:', word_counter_1941.most_common(3))
print('Top 3 word in 1961-Kennedy:', word_counter_1961.most_common(3))
print('Top 3 word in 1973-Nixon:', word_counter_1973.most_common(3))
```
